# 2-Preprocessing Agg Data

### Import & Turn the Fractions into 1 & 5 minutes Candles of True & False Buyer Makers + Vdiff

```python
import os
import pandas as pd
import numpy as np

folder_path = r"C:\Users\Mcc\Desktop\Data\Historical Data\Agg Volumes\BTCUSDT-aggTrades-2024-04-01 - 2024-04-17\COIN-M"
column_names = ["agg_trade_id", "price", "quantity", "first_trade_id", "last_trade_id", "transact_time", "is_buyer_maker"]

def remove_milliseconds(datetime_str):
    return datetime_str.strftime('%Y-%m-%d %H:%M:%S')

def generate_minute_volumes(df_agg, is_buyer_maker):
    df_filtered = df_agg.loc[df_agg['is_buyer_maker'] == is_buyer_maker]
    df_minute_volumes = df_filtered.resample('5T')['volume'].sum().fillna(0)
    #print(df_minute_volumes)
    return df_minute_volumes

for file in os.listdir(folder_path):
    if file.endswith('.csv'):  # Check if the file is a CSV file
        file_path = os.path.join(folder_path, file)  # Get the full file path

        #df_agg = pd.read_csv(file_path, header=None, names=column_names)
        df_agg = pd.read_csv(file_path)
        df_agg['transact_time'] = pd.to_datetime(df_agg['transact_time'], unit='ms')
        df_agg.set_index('transact_time', inplace=True)
        df_agg['volume'] = df_agg['price'] * df_agg['quantity']

        start_time_str = df_agg.index[0]
        end_time_str = df_agg.index[-1]

        df_maker_true = generate_minute_volumes(df_agg, True)
        df_maker_false = generate_minute_volumes(df_agg, False)

        merged_df = pd.merge(df_maker_true, df_maker_false, left_index=True, right_index=True, how='outer', suffixes=('_true', '_false'))
        merged_df['vdiff(false-true)'] = merged_df['volume_false'] - merged_df['volume_true']
        #merged_df.index = merged_df.index.floor('s')

        # Extract filename without extension
        filename = os.path.splitext(file)[0]
        new_filename = f"{filename}-COINM-5M.csv"  # New CSV filename
        merged_df.to_csv(new_filename)
        current_time = pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"Finished & Saved - {new_filename} - at {current_time}")
```

### Resample to 1 & 5 Minutes

```python
import pandas as pd

# Assuming df is your DataFrame with the funding rates
# Assuming calc_time is the index of your DataFrame

# Resample to 1-minute frequency and forward fill missing values
df_resampled = df_fund.resample('1T').ffill()

# Generate the complete date range from the start to the end of your desired period
start_date = '2020-01-01 00:00:00.000'
end_date = '2024-03-31 16:00:00.001'
date_range = pd.date_range(start=start_date, end=end_date, freq='5T')

# Reindex the DataFrame with the complete date range
df_complete = df_resampled.reindex(date_range)

# Fill any remaining missing values with the last available value
df_complete = df_complete.ffill()

# If needed, reset index to have 'calc_time' as a column instead of index
df_complete.reset_index(inplace=True)

# Rename the columns if needed
# df_complete = df_complete.rename(columns={'index': 'calc_time'})

# Now df_complete contains funding rates at 1-minute frequency from 2020-01-01 to 2024-03-31 16:00:00.001
```
